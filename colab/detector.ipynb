# Descargar el resultado ✅

```python
import os
import glob
from google.colab import files

output_dir_base = "runs/detect"
output_video_filename = "choque.mp4" #Colocar nombre del video

# Enumera todos los directorios 'exp' y encuentra el que tenga la fecha de modificación más reciente.
all_exp_dirs = sorted(glob.glob(os.path.join(output_dir_base, "exp*")),
                      key=os.path.getmtime, reverse=True)

found_video_path = None
for exp_dir in all_exp_dirs:
    potential_path = os.path.join(exp_dir, output_video_filename)
    if os.path.exists(potential_path):
        found_video_path = potential_path
        break

if found_video_path:
    # Descarga del video
    print(f"Descargando video: {found_video_path}")
    files.download(found_video_path)
else:
    print(f"No se encontró el video '{output_video_filename}' en ninguna de las carpetas '{output_dir_base}/exp*'.")
    print("Contenido de runs/detect:")
    !ls -R {output_dir_base}
```
